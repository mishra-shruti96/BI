```python
#showing correlation plot between different features in the dataset.

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import seaborn as sns
#Seaborn is a library for making statistical graphics in Python.
#It is built on top of matplotlib and closely integrated with pandas data structures.

df= pd.read_csv('forestfires.csv')
#data = pd.read_csv('forestfires.csv', usecols=['X','Y','month','day','FFMC','DMC','DC','ISI','temp','RH','wind','rain','area'])
names = ['X','Y','month','day','FFMC','DMC','DC','ISI','temp','RH','wind','rain','area','fire']
data=df[df.columns[0:15]]
corr=data.corr()
plt.subplots(figsize=(15,15))
sns.heatmap(corr,annot = True, vmin=-1, vmax=1, center= 0, cmap= 'coolwarm',square=True)

#scatter plot
scatter_matrix(data,figsize=[20,20])
plt.show()
```
